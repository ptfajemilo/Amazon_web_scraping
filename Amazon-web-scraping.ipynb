```python
#obtaining the content as a website html file
with open('webpage.html', 'w', encoding = "utf-8") as f:
    f.write(page_contents)
```

```python
#using the beautiful soup to pretify the contents

doc = BeautifulSoup(page_contents, "html.parser")
print(doc)
```

```python
#Extracting the span tag and the  tag. The p tag for the title and description while the a is for the link
label_class = "a-size-base-plus a-color-base a-text-normal"
label_tags = doc.find_all('span',class_ = label_class )

rating_selector = 'a-size-base'
rating_tags = doc.find_all('span', class_=rating_selector)

price_selector  = 'a-price-whole'
price_tags = doc.find_all('span', class_= price_selector)

best_seller_selector = 'a-badge-text'
best_seller_tag =  doc.find_all('span', class_= best_seller_selector)

link_selector = 'a-link-normal s-no-outline'
link_tags = doc.find_all('a', class_= link_selector)
```

```python
#Extracting the exact text and appending in a list
labels = []
for label in label_tags:
    labels.append(label.text)
```
